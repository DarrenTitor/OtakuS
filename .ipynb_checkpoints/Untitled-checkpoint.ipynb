```python
df = spark.read.csv('archive/anime_cleaned.csv',
                    header='true', 
                    inferSchema='true')
```

```python
df2 = spark.read.csv('archive/users_cleaned.csv',
                    header='true', 
                    inferSchema='true')
```

```python
anime_id_list = df.select('anime_id').sample(0.15, seed=0).collect()
anime_id_set = set([row.anime_id for row in anime_id_list])
del anime_id_list
```

```python
len(anime_id_set)
```

```python
df.columns
```

```python
useful_columns = ['type','episodes','status','airing','score','scored_by','rank', 'popularity','members','favorites','genre',  'anime_id_vector',]
```

```python
useless_columns = [col for col in df.columns if col not in useful_columns]
```

```python
df = df.drop(*useless_columns)
```

```python
df.show(1, vertical=True)
```

```python
user_id_list = df2.select('user_id').sample(0.03, seed=0).collect()
user_id_set = set([row.user_id for row in user_id_list])
del user_id_list
```

```python
len(user_id_set)
```

```python
from pyspark.ml.linalg import Vectors
```

```python
# One hot the anime_id
ohe = pyspark.ml.feature.OneHotEncoder()
ohe.setInputCols(["anime_id"])
ohe.setOutputCols(["anime_id_vector"])
ohe_model = ohe.fit(df)
ohe_model.setOutputCols(["anime_id_vector"])
ohe_model.getHandleInvalid()
df = ohe_model.transform(df)
```
